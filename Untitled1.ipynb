# 柔軟モデルのテスト

```python
from collections import deque
import itertools

def get_packet_swap_neighbors(state):
    """
    任意の2つの互いに素なパケットを入れ替える全パターンを生成する
    """
    n = len(state)
    neighbors = []

    # パケット1の範囲 (iからj-1)
    for i in range(n):
        for j in range(i + 1, n + 1):
            p1_range = (i, j)

            # パケット2の範囲 (kからl-1)
            # パケット1と重ならない範囲を探す
            for k in range(n):
                for l in range(k + 1, n + 1):
                    p2_range = (k, l)

                    # 互いに素である判定 (p1の終わり <= p2の開始 or p2の終わり <= p1の開始)
                    if j <= k or l <= i:
                        # 入れ替え処理
                        # 非破壊的にリストを操作
                        res = list(state)
                        idx1 = list(range(i, j))
                        idx2 = list(range(k, l))

                        # 物理的なパケット交換（中身を入れ替えるのではなく、位置を入れ替える）
                        # 1. 小さいインデックスの方を先に特定
                        first_range = idx1 if i < k else idx2
                        second_range = idx2 if i < k else idx1

                        # 2. 3つのパーツに分割して再構成（[前][中][後]）
                        # ※ 簡易化のため、ここでは単純な位置交換をシミュレート
                        new_state = list(state)
                        p1_content = state[i:j]
                        p2_content = state[k:l]

                        # 単純な要素のスワップではなく、パケット単位の移動を模した処理
                        # (実装の詳細は物理的な「パケット交換」の定義に基づく)
                        # ここでは最も広義の「2つのパケットの位置を入れ替える」を実装
                        if j <= k: # p1が前にある場合
                            new_state = state[:i] + p2_content + state[j:k] + p1_content + state[l:]
                        else: # p2が前にある場合
                            new_state = state[:k] + p1_content + state[l:i] + p2_content + state[j:]

                        neighbors.append(tuple(new_state))
    return set(neighbors)

def solve_min_cost(start_node, target_node):
    """
    BFSによる最短経路探索
    """
    if start_node == target_node:
        return 0

    queue = deque([(start_node, 0)])
    visited = {start_node}

    while queue:
        current, dist = queue.popleft()

        for neighbor in get_packet_swap_neighbors(current):
            if neighbor == target_node:
                return dist + 1
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, dist + 1))
    return -1

# --- 検証実行 ---
n = 5
start = tuple(range(1, n + 1))  # (1, 2, 3, 4, 5)

# テストケース1: 巡回シフト (Cut) -> コスト1になるはず
target_shift = (2, 3, 4, 5, 1)
print(f"Shift Cost: {solve_min_cost(start, target_shift)}")

# テストケース2: 離れた単一カードの交換 (1,3) -> コスト1になるはず
target_swap = (3, 2, 1, 4, 5)
print(f"AnySwap Cost: {solve_min_cost(start, target_swap)}")

# テストケース3: 複雑な置換
target_complex = (2, 4, 1, 3, 5)
print(f"Complex Cost: {solve_min_cost(start, target_complex)}")
```

### 全置換の場合のシャッフルとの比較

```python
from collections import deque
import matplotlib.pyplot as plt

# --- 1. 探索用関数 (以前の定義) ---
def get_packet_swap_neighbors(state):
    """
    柔軟モデル：任意の2つの互いに素なパケットを入れ替える
    """
    n = len(state)
    neighbors = []

    # パケット1の範囲 (iからj-1)
    for i in range(n):
        for j in range(i + 1, n + 1):
            # パケット2の範囲 (kからl-1)
            for k in range(n):
                for l in range(k + 1, n + 1):
                    # 互いに素である判定
                    if j <= k or l <= i:
                        # パケット交換処理
                        p1_content = state[i:j]
                        p2_content = state[k:l]

                        if j <= k: # p1が前
                            new_state = state[:i] + p2_content + state[j:k] + p1_content + state[l:]
                        else: # p2が前
                            new_state = state[:k] + p1_content + state[l:i] + p2_content + state[j:]

                        neighbors.append(tuple(new_state))
    return set(neighbors)

# --- 2. 統計計算関数 (ご提示のコード) ---
def calculate_all_permutation_stats(n):
    start = tuple(range(1, n + 1))
    queue = deque([(start, 0)])
    all_costs = {start: 0}

    while queue:
        current, dist = queue.popleft()
        for neighbor in get_packet_swap_neighbors(current):
            if neighbor not in all_costs:
                all_costs[neighbor] = dist + 1
                queue.append((neighbor, dist + 1))

    stats = {}
    for cost in all_costs.values():
        stats[cost] = stats.get(cost, 0) + 1

    total_permutations = len(all_costs)
    avg_cost = sum(all_costs.values()) / total_permutations

    return stats, avg_cost, total_permutations

# --- 3. グラフ描画関数 (新規追加) ---
def plot_distribution(stats, n, avg_cost):
    """
    コスト分布を棒グラフで描画し、数値をラベル表示する
    """
    costs = sorted(stats.keys())
    counts = [stats[c] for c in costs]
    total = sum(counts)

    # グラフの設定
    plt.figure(figsize=(10, 6))
    bars = plt.bar(costs, counts, color='skyblue', edgecolor='black', zorder=3)

    # デザイン調整
    plt.title(f'Cost Distribution of Permutations (n={n}, Flexible Model)', fontsize=16)
    plt.xlabel('Cost (Steps)', fontsize=14)
    plt.ylabel('Number of Permutations', fontsize=14)
    plt.xticks(costs)  # X軸を整数のみに
    plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)

    # 棒の上に数値とパーセンテージを表示
    for bar in bars:
        height = bar.get_height()
        percentage = (height / total) * 100
        plt.text(bar.get_x() + bar.get_width()/2, height + 1,
                 f'{height}\n({percentage:.1f}%)',
                 ha='center', va='bottom', fontsize=12, fontweight='bold')

    # 平均コストなどをテキストボックスで表示
    text_str = f'Total States: {total}\nAverage Cost: {avg_cost:.3f}'
    plt.text(0.95, 0.95, text_str, transform=plt.gca().transAxes,
             fontsize=12, verticalalignment='top', horizontalalignment='right',
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

    plt.tight_layout()

    # 画像として保存 (スライド貼付用)
    filename = f"permutation_cost_dist_n{n}.png"
    plt.savefig(filename, dpi=300)
    print(f"Graph saved as: {filename}")
    plt.show()

# --- 実行 ---
n = 5
print("Calculating... (This may take a moment)")
distribution, average, total = calculate_all_permutation_stats(n)

print(f"\n--- Statistics for n={n} (Flexible Model) ---")
print(f"Total Permutations: {total}")
for cost, count in sorted(distribution.items()):
    print(f"Cost {cost}: {count} states ({count/total*100:.1f}%)")
print(f"Average Cost: {average:.3f}")

# グラフ描画の実行
plot_distribution(distribution, n, average)
```

# 厳格モデルのテスト

```python
from collections import deque

def get_strict_neighbors(state):
    """
    厳格モデルの近傍生成:
    1. Cut: 任意の箇所での巡回シフト (コスト1)
    2. Adjacent Swap: 隣接する2枚の入れ替え (コスト1)
    """
    n = len(state)
    neighbors = set()

    # 1. Cut (巡回シフト): 全てのシフト量を生成元とする
    for k in range(1, n):
        # 左にk個ずらす
        new_state = state[k:] + state[:k]
        neighbors.add(tuple(new_state))

    # 2. Adjacent Swap (隣接互換): 隣り合うペアのみ交換
    for i in range(n - 1):
        new_state = list(state)
        new_state[i], new_state[i+1] = new_state[i+1], new_state[i]
        neighbors.add(tuple(new_state))

    return neighbors

def solve_strict_cost(start_node, target_node):
    if start_node == target_node:
        return 0

    queue = deque([(start_node, 0)])
    visited = {start_node}

    while queue:
        current, dist = queue.popleft()
        for neighbor in get_strict_neighbors(current):
            if neighbor == target_node:
                return dist + 1
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, dist + 1))
    return -1

# --- 検証実行 ---
n = 5
start = tuple(range(1, n + 1)) # (1, 2, 3, 4, 5)

# テストケース1: 巡回シフト (2, 3, 4, 5, 1)
target_shift = (2, 3, 4, 5, 1)
# テストケース2: 離れた単一カードの交換 (3, 2, 1, 4, 5)
target_swap = (3, 2, 1, 4, 5)
# テストケース3: 複雑な置換 (2, 4, 1, 3, 5)
target_complex = (2, 4, 1, 3, 5)

print(f"Shift Cost (Strict): {solve_strict_cost(start, target_shift)}")
print(f"AnySwap Cost (Strict): {solve_strict_cost(start, target_swap)}")
print(f"Complex Cost (Strict): {solve_strict_cost(start, target_complex)}")
```
